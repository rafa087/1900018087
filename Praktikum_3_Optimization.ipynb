##**1. Pengenalan Optimizer**


Optimasi (optimization), adalah hal yang krusial dan penting pada semua model Machine Learning. Setiap model Machine Learning yang ditraining pada data training pada dasarnya mencoba untuk meminimalkan nilai *loss function*. Tujuan utama pelatihan/training model Neural Network adalah untuk mendapatkan model yang memiliki nilai "loss" yang paling kecil. Optimizer merupakan bagian yang esensial yang mem-*fine-tune* parameter-parameter Neural Network untuk mengurangi  perbedaan antara nilai output asli dengan nilai output hasil prediksi.

Terdapat beberapa algoritma optimasi yang digunakan untuk meminimalkan *loss function* pada pelatihan Neural Network. Pada praktikum ini, Anda akan mencoba menggunakan 3 jenis Optimizer yaitu: **Stochastic Gradient Descent** (SGD), **RMSProp**, dan **Adam** (Adaptive Moment Optimizer).

## Stochastic Gradient Descent (SGD)


Hampir semua loss function yang digunakan pada suatu model Machine Learning mengandung semua fungsi penjumlahan yang melibatkan semua data training, sebagai contoh, mean squared error:


$$f(w)=\frac{1}{n}\sum^{n}_{i=1}(h_w(x_i)-y_i)^2$$

Disini $f(w)$ adalah nilai dari loss function, $h_w(x)$ adalah model yang kita bangun, sebagai contoh untuk linear regression: $h_w(x)=wTx$. Sedangkan $y$ adalah nilai target yang diharapkan. Tujuan utamanya adalah mencari nilai $w$ yang meminimalkan nilai loss function. Dengan gradient descent kita lakukan:

    Mulai dengan nilai w yang random
    Hitung gradient menggunakan semua data pada training set
    Gunakan gradient untuk menghitung ulang nilai w
    Ulangi sebanyak I iterations atau sampai step-size nya kecil

Jika kita memiliki data yang besar (n besar) gradient descent sangat computational costly (mahal secara komputasi).
Kita dapat mengurangi beban komputasi dengan hanya menggunakan porsi/sebagian dari data pada training set untuk  mengupdate parameter pada tiap iterasi (daripada mnenggunakan semua data pada training set). Pendekatan ini disebut "stochastic gradient descent".

**Gradient Descent**:

$$w_{j+1}=w_{j}-\alpha_t\frac{\partial}{\partial w_j}f(w_j)$$

**Stochastic Gradient Descent**:

$$w_{j+1}=w_{j}-\alpha_t\frac{\partial}{\partial w_j}f_i(w_j)$$

Apa perbedaan kedua rumus di atas?
Pada Stochastic Gradient Descent fungsi $f$ memiliki subscript $i$. Artinya, daripada memperbarui parameter kita berdasarkan gradien yang dihitung menggunakan semua data training, kita cukup menggunakan salah satu titik data  (yaitu data yang ke i)

## RMSProp

Hinton (2012) mengusulkan algoritma RMSProp algorithm sebagai perbaikan sederhana untuk memisahkan penjadwalan laju dari coordinate-adaptive learning rates. RMSProp sebenarnya termotivasi dari algoritma optimasi Adagrad. Permasalahannya adalah pada optimasi menggunakan algoritma Adagrad kuadrat gradien $\mathbf{g}_t$ diakumulasi menjadi sebuah vektor $\mathbf{s}_t = \mathbf{s}_{t-1} + \mathbf{g}_t^2$. Hasilnya, nilai $\mathbf{s}_t$ tumbuh tanpa batasan karena tidak ada normalisasi.

Salah satu cara untuk menyelesaikan problem tersebut adalah menggunakan $\mathbf{s}_t / t$. Untuk distribusi yang 'reasonable' nilai $\mathbf{g}_t$ akan menjadi konvergen. Sayangnya, mungkin butuh waktu yang sangat lama hingga pemberlakukan batas menjadi penting karena prosedurnya mengingat lintasan nilai-nilai secara penuh.

Alternatif lainnya adalah memakai *leaky average* seperti halnya digunakan pada metode Momentum, misalnya, $\mathbf{s}_t \leftarrow \gamma \mathbf{s}_{t-1} + (1-\gamma) \mathbf{g}_t^2$ untuk beberapa parameter nilai $\gamma > 0$. Ketika semua bagian tidak mengalami perubahan, maka dihasilkanlah RMSProp.

## Algoritma

Persamaan detail:

$$\begin{aligned}
    \mathbf{s}_t & \leftarrow \gamma \mathbf{s}_{t-1} + (1 - \gamma) \mathbf{g}_t^2, \\
    \mathbf{x}_t & \leftarrow \mathbf{x}_{t-1} - \frac{\eta}{\sqrt{\mathbf{s}_t + \epsilon}} \odot \mathbf{g}_t.
\end{aligned}$$

Konstanta $\epsilon > 0$ biasanya di-set menjadi $10^{-6}$ untuk memastikan tidak ada pembagian dengan 0 atau nilai *step size* yang terlalu besar. Dengan ekspansi ini kita dapat mengontrol laju *learning rate* $\eta$ secara independen dari scaling yang diaplikasikan per-basis koordinat. Dalam hal leaky averages kita dapat mengaplikasikan alasan yang serupa seperti yang diaplikasikan pada momentum. Ekspansi definisi of $\mathbf{s}_t$ menghasilkan:

$$
\begin{aligned}
\mathbf{s}_t & = (1 - \gamma) \mathbf{g}_t^2 + \gamma \mathbf{s}_{t-1} \\
& = (1 - \gamma) \left(\mathbf{g}_t^2 + \gamma \mathbf{g}_{t-1}^2 + \gamma^2 \mathbf{g}_{t-2} + \ldots, \right).
\end{aligned}
$$

Kita gunakan  $1 + \gamma + \gamma^2 + \ldots, = \frac{1}{1-\gamma}$. Sehingga jumlah dari bobot dinormalisasi menjadi $1$ waktu paruh pengamatan $\gamma^{-1}$.

## Adam Optimizer

Adam (singkatan Adaptive Moment Estimation) merupakan optimizer yang menggabungkan kekuatan dua tenik optimasi, yaitu Momentum dan RMSprop, untuk mengatur *learning rate* pada pelatihan Neural Network.

Adam optimzer sangat efektif, terutama jika bekerja dengan data yang besar dan arsitektur Neural network yang komplex, karena ia efisien dalam penggunaan memory, dan mengatur *learning rate* secara dinamis untuk setiap parameter Network.

Dua konsep kunci pada Adam Optimizer yaitu: Momentum dan RMSProp


Salah satu komponen kunci dari optimasi Adam ia menggunakan nilai rata-rata bergerak berbobot yang eksponensial (also known as leaky averaging, atau EMWA- *exponentially moving weighted averages*) untuk mendapatkan estimasi nilai momentum dan gradien momen. Ia menggunakan variabel-variabel state:

$$\begin{aligned}
    \mathbf{v}_t & \leftarrow \beta_1 \mathbf{v}_{t-1} + (1 - \beta_1) \mathbf{g}_t, \\
    \mathbf{s}_t & \leftarrow \beta_2 \mathbf{s}_{t-1} + (1 - \beta_2) \mathbf{g}_t^2.
\end{aligned}$$

Di sini $\beta_1$ dan $\beta_2$ parameter bobot yang non-negatif. Umumnya pilihan nilainya adalah $\beta_1 = 0.9$ and $\beta_2 = 0.999$. Variansi nilai estimasi bergerak *much more slowly* (jauh lebih lambat) dibandingkan momentumnya. Perlu dicatat jika kita inisialisasi $\mathbf{v}_0 = \mathbf{s}_0 = 0$ kita memiliki nilai bias yang sangat signifikan  terhadap nilai-nilai yang kecil. Hal ini dapat diselesaikan dengan menggunakan $\sum_{i=0}^{t-1} \beta^i = \frac{1 - \beta^t}{1 - \beta}$ untuk me-renormalisasi. Setelah dinormalisasi, variabel-variabel state nya menjadi:

$$\hat{\mathbf{v}}_t = \frac{\mathbf{v}_t}{1 - \beta_1^t} \textrm{ and } \hat{\mathbf{s}}_t = \frac{\mathbf{s}_t}{1 - \beta_2^t}.$$

Berbekal estimasi yang tepat, kita sekarang dapat menuliskan update persamaannya. Pertama, kita mengubah skala gradien dengan cara yang sangat mirip dengan RMSProp untuk memperoleh:

$$\mathbf{g}_t' = \frac{\eta \hat{\mathbf{v}}_t}{\sqrt{\hat{\mathbf{s}}_t} + \epsilon}.$$

Tidak seperti RMSProp, update pada Adam menggunakan momentum $\hat{\mathbf{v}}_t$ daripada gradien itu sendiri. Selain itu, terdapat sedikit perbedaan sebab rescaling dilakukan menggunakan $\frac{1}{\sqrt{\hat{\mathbf{s}}_t} + \epsilon}$ dan bukan $\frac{1}{\sqrt{\hat{\mathbf{s}}_t + \epsilon}}$. Rumus pertama bekerja dengan lebih baik, oleh karena itu terdapat perbedaan dengan RMSProp. Biasanya $\epsilon = 10^{-6}$ digunakan karena memiliki trade-off antara stabilitas numerik dan ketepatan.

Sekarang kita memiliki semua bagian yang diperlukan untuk menghitung update. Ini sedikit antiklimaks dan kita memiliki update sederhana dalam bentuk:

$$\mathbf{x}_t \leftarrow \mathbf{x}_{t-1} - \mathbf{g}_t'.$$

##**3. Import Library**

```python
# Importing necessary modules
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Flatten, Dense
```

## **4. Download Dataset**

Pada praktikum ini kita gunakan dataset MNIST yang berisi gambar-gambar angka tulisan tangan. Gambar-gambar tersebut menunjukkan angka 1 sampai angka 9. Klasifikasi dapat dilakukan untuk memberi label pada gambar sesuai dengan angka yang ditampilkan. Misalnya Gambar yang memuat angka satu diberikan label '1' dan Gambar yang memuat angka enam diberikan label '6'

Setelah load data MNIST, dilakukan normalisasi pada data gambar dengan membagi dengan 255 (karena nilai piksel berkisar dari 0 hingga 255), yang membantu konvergensi yang lebih cepat selama pelatihan.

```python
# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalisasi image dengan membagi dengan 255 agar gambar menjadi Gray Scale
gray_scale = 255

x_train = x_train.astype('float32') / gray_scale
x_test = x_test.astype('float32') / gray_scale

# Mengecek shape dari data training dan data testing (fitur dan target nya)
print("Feature matrix (x_train):", x_train.shape)
print("Target matrix (y_train):", y_train.shape)
print("Feature matrix (x_test):", x_test.shape)
print("Target matrix (y_test):", y_test.shape)
```

##**5.Membangun Model Jaringan Syarat Tiruan (Multi Layer Perceptron) dengan Optimizer SGD**

```python
# Building the Sequential neural network model
model1 = Sequential([
    Flatten(input_shape=(28, 28)),     #ini input layer yang sekaligus Flatten
    Dense(256, activation='sigmoid'),   #ini hidden layer pertama
    Dense(128, activation='sigmoid'),   #ini hidden layer kedua
    Dense(64, activation='sigmoid'),  #ini  hidden layer ketiga
    Dense(10, activation='softmax'),   #ini output layer dengan fungsi aktivasi Softmax untuk multi-class classification
])
```

Optimizer di-set ketika kita memanggil model.compile
Pada percobaan pertama kita akan gunakan Optimizer Stochastic Gradient Descent



```python
model1.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.001, momentum=0.1),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

Training model dengan memanggil model.fit. Gunakan epochs 10 dan validation split 0.2

```python
history = model1.fit(x_train, y_train, epochs=10,
          batch_size=2000,
          validation_split=0.2)
```

Setelah model di-training, evaluasi performa model dengan data test. Catatlah akurasi yang diperoleh ketika Optimizer yang digunakan adalah SGD.

```python


results = model1.evaluate(x_test, y_test, verbose=0)
print('Test loss, Test accuracy:', results)
```

Simpan model dalam bentuk file h5

```python
model1.save('gdrive/My Drive/Colab Notebooks/Deep Learning Labs/Prak03_Data/OptimizerSGD.h5')
```
